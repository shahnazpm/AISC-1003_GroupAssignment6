```python
# Identify features with more than 60% missing values
features_to_drop = null_percentage[null_percentage > 60].index

# Drop the identified features from the dataset
df_cleaned = df.drop(columns=features_to_drop)

# Display the cleaned dataframe
print("Cleaned DataFrame Preview:")
print(df_cleaned.head())

# Print the number of features removed
print(f"\nRemoved {len(features_to_drop)} features with more than 60% null values")

# Print the features that were dropped
print("\nFeatures dropped due to >60% missing values:")
for feature in features_to_drop:
    print(f"- {feature}")
```

```python
# Function to plot correlation heatmap for a subset of columns
def plot_correlation_heatmap(df, columns, title='Correlation Matrix Heatmap'):
    correlation_matrix = df[columns].corr()
    plt.figure(figsize=(12, 10))
    sns.heatmap(correlation_matrix, annot=True, cmap='RdBu', vmin=-1, vmax=1)
    plt.title(title)
    plt.show()

# Select numeric columns
numeric_df_cleaned = df_cleaned.select_dtypes(include=['int64', 'float64'])

# Split numeric columns into subsets
columns = numeric_df_cleaned.columns
num_features = len(columns)
subset_size = 15  # Number of features per subset

# Plot correlation heatmaps for each subset
for start in range(0, num_features, subset_size):
    end = min(start + subset_size, num_features)
    subset_columns = columns[start:end]
    plot_correlation_heatmap(numeric_df_cleaned, subset_columns, title=f'Correlation Matrix Heatmap: Features {start + 1} to {end}')
```

```python
# Calculate correlation matrix
correlation_matrix = numeric_df_cleaned.corr()

# Get the upper triangle of the correlation matrix
upper_triangle = correlation_matrix.where(np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool))

# Find index of feature columns with correlation greater than a threshold (e.g., 0.75)
highly_correlated = [(col, idx, correlation_matrix.loc[col, idx]) for col in upper_triangle.columns for idx in upper_triangle.index if upper_triangle.loc[idx, col] > 0.5]

# Convert to DataFrame for better readability
high_corr_df = pd.DataFrame(highly_correlated, columns=['Feature1', 'Feature2', 'Correlation'])

# Display top correlated features
print(high_corr_df.sort_values(by='Correlation', ascending=False))
```
